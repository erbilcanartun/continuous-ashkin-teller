## Multiple RG Flows

```python
mp.dps = 50

J_values = [.2, .3, .35,.37,.39,.4,.42,.44,.46,.47,.5,.55,.6]

b = 3
d = 2

n_max = 50
n_steps = 70

for i, J in enumerate(J_values):
    M = J
    # Track RG flow
    flow_history = track_rg_flow(J, M, b, d, n_max, n_steps)
    
    print(f"J = {J}, M = {M}\n")
    coefficient_sink(flow_history, rg_step=-1)
    
    # Coefficient evolution
    fig = plot_coefficient_evolution(flow_history, J, M)
    plt.show()
    folder = "/Users/artun/Desktop/"
    fig.savefig(folder + f"coefficient_evolution_J{J}M{M}.png")
```
